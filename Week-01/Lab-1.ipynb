```python
#1
rows = 7
columns = 9

array_2d = []
for i in range(rows):
    row = []
    for j in range(columns):
        value = int(input(f"Enter value for row {i + 1}, column {j + 1}: "))
        row.append(value)
    array_2d.append(row)

print("The 2D array is:")
for row in array_2d:
    print(row)
```

```python
#2

my_dict = {}

for i in range(5):
  key = input(f"Enter key {i + 1}: ")
  values = []
  for j in range(10):
    value_type = input(f"Enter type for value {j + 1} (string, integer, float): ").lower()
    if value_type == "string":
      value = input(f"Enter string value: ")
    elif value_type == "integer":
      value = int(input(f"Enter integer value: "))
    elif value_type == "float":
      value = float(input(f"Enter float value: "))
    else:
      print("Invalid value type. Please enter string, integer, or float.")
      j -= 1
      continue
    values.append(value)
  my_dict[key] = values

print("The dictionary is:", my_dict)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


data = {'Category': ['A', 'B', 'A', 'C', 'B']}
df = pd.DataFrame(data)

label_encoder = LabelEncoder()

df['Category_Encoded'] = label_encoder.fit_transform(df['Category'])

print(df)
```
