## Week-02 Task

1.Show BFS on this graph.
2.Find out (A,D,H,I)

```python
graph = {
    'A': {'B': 10, 'C': 3, 'D': 1},
    'B': {'E': 11, 'F': 4},
    'C': {'G': 4},
    'D': {'G': 6, 'H': 1},
    'E': {'I': 3},
    'F': {'C': 9, 'I': 6},
    'G': {'H': 7},
    'H': {'I': 1},
    'I': {}
}
```

```python

visited = []
queue = []
def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        s = queue.pop(0)
        print(s, end=" ")
        for neighbour, weight in graph[s].items():
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
                print("Weight:", weight)

print("BFS:")
bfs(visited, graph, 'A', 'D', 'H', 'I')
```
