```python
#BFS
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}
visited = []  # list for visited nodes
queue = []    # Initilize a queue
```

```python
def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:  # creating loop to visit  each node
        s = queue.pop(0)
        print(s, end=" ")
        for neighbour in graph[s]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
print("Following is the Breadth-First Search")
bfs(visited, graph, '5')
```

```python
# prompt: make this sort for human: from collections import deque
# # Graph with weights
# graph = {
#     'A': [('B', 10), ('C', 3)],
#     'B': [('E', 11)],
#     'C': [('D', 4), ('F', 9)],
#     'D': [('H', 1)],
#     'E': [('I', 3)],
#     'F': [('I', 6), ('H', 1)],
#     'H': [('I', 1)],
#     'I': []
# }
# # BFS Traversal
# def bfs(graph, start):
#     queue = deque([start])
#     visited = set()
#     traversal_order = []
#     while queue:
#         node = queue.popleft()
#         if node not in visited:
#             visited.add(node)
#             traversal_order.append(node)
#             for neighbor, _ in graph[node]:
#                 queue.append(neighbor)
#     return traversal_order
# # Find minimum cost path using simple path accumulation
# def find_min_cost_path(graph, start, path_nodes):
#     total_cost = 0
#     path = [start]
#     for target in path_nodes:
#         cost, sub_path = bfs_shortest_path(graph, start, target)
#         total_cost += cost
#         path.extend(sub_path[1:])
#         start = target
#     return total_cost, path
# # Helper for shortest path using BFS (without priority queue for simplicity)
# def bfs_shortest_path(graph, start, target):
#     queue = deque([(start, 0, [start])])  # (current_node, cost, path)
#     while queue:
#         node, cost, path = queue.popleft()
#         if node == target:
#             return cost, path
#         for neighbor, weight in graph[node]:
#             if neighbor not in path:  # Avoid revisiting nodes
#                 queue.append((neighbor, cost + weight, path + [neighbor]))
#     return float('inf'), []
# # Execute BFS
# print("BFS Traversal:", bfs(graph, 'A'))
# # Find shortest path with minimum cost for A -> D -> H -> I
# path_nodes = ['D', 'H', 'I']
# cost, path = find_min_cost_path(graph, 'A', path_nodes)
# print("Minimum Cost Path from A -> D -> H -> I:", path)
# print("Total Cost:", cost)

from collections import deque

# Graph with weights
graph = {
    'A': [('B', 10), ('C', 3)],
    'B': [('E', 11)],
    'C': [('D', 4), ('F', 9)],
    'D': [('H', 1)],
    'E': [('I', 3)],
    'F': [('I', 6), ('H', 1)],
    'H': [('I', 1)],
    'I': []
}

# BFS Traversal
def bfs(graph, start):
    queue = deque([start])
    visited = set()
    traversal_order = []
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            traversal_order.append(node)
            for neighbor, _ in graph[node]:
                queue.append(neighbor)
    return traversal_order

# Find minimum cost path using simple path accumulation
def find_min_cost_path(graph, start, path_nodes):
    total_cost = 0
    path = [start]
    for target in path_nodes:
        cost, sub_path = bfs_shortest_path(graph, start, target)
        total_cost += cost
        path.extend(sub_path[1:])
        start = target
    return total_cost, path

# Helper for shortest path using BFS (without priority queue for simplicity)
def bfs_shortest_path(graph, start, target):
    queue = deque([(start, 0, [start])])  # (current_node, cost, path)
    while queue:
        node, cost, path = queue.popleft()
        if node == target:
            return cost, path
        for neighbor, weight in graph[node]:
            if neighbor not in path:  # Avoid revisiting nodes
                queue.append((neighbor, cost + weight, path + [neighbor]))
    return float('inf'), []

# Execute BFS
print("BFS Traversal:", bfs(graph, 'A'))

# Find shortest path with minimum cost for A -> D -> H -> I
path_nodes = ['D', 'H', 'I']
cost, path = find_min_cost_path(graph, 'A', path_nodes)
print("Minimum Cost Path from A -> D -> H -> I:", path)
print("Total Cost:", cost)
```
