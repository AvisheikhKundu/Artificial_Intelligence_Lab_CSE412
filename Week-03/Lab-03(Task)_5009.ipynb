## DFS

```python
# DFS- iterative method
```

```python
graph = {
    '2': ['4', '3'],
    '4': ['8', '7'],
    '3': ['5'],
    '8': [],
    '7': [],
    '5': []
}

def dfs_iterative(graph, start):
    visited = set()
    stack = [start]

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=", ")
            visited.add(node)
            stack.extend(reversed(graph[node]))

print("dfs:")
dfs_iterative(graph, '2')
```

# Water Jug Problem

```python
from collections import defaultdict
jug1, jug2, aim = 4, 3, 2
visited = defaultdict(lambda: False)
def waterJugSolver(amt1, amt2):
    if (amt1 == aim and amt2 == 0) or (amt2 == aim and amt1 == 0):
        print(amt1, amt2)
        return True
    if visited[(amt1, amt2)] == False:
        print(amt1, amt2)
        visited[(amt1, amt2)] = True
        return (waterJugSolver(0, amt2) or
                waterJugSolver(amt1, 0) or
                waterJugSolver(jug1, amt2) or
                waterJugSolver(amt1, jug2) or
                waterJugSolver(amt1 + min(amt2, (jug1-amt1)),
                amt2 - min(amt2, (jug1-amt1))) or
                waterJugSolver(amt1 - min(amt1, (jug2-amt2)),
                amt2 + min(amt1, (jug2-amt2))))
    else:
        return False
print("Steps to reach solution: ")
if not waterJugSolver(0, 0):
    print("No solution")
```

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()  # Initialize the set to keep track of visited nodes

    visited.add(start)
    print(start, end=" ")  # Print the current node

    # Visit all the neighbors of the current node that haven't been visited yet
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Example usage:
graph = {
    '2': ['4', '3'],  # Visit 4 first, then 3
    '4': ['8', '7'],  # Visit 8 first, then 7
    '3': ['5'],       # Visit 5 last
    '8': [],          # No further neighbors for 8
    '7': [],          # No further neighbors for 7
    '5': []           # No further neighbors for 5
}

print("Following is the DFS (Recursive):")
dfs(graph, '2')
```
