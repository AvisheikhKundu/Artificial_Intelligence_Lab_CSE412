```python
import math
from collections import deque

def canMeasureWater(jug1, jug2, target):
    if target > jug1 + jug2 or target % math.gcd(jug1, jug2) != 0:
        return False

    queue = deque([(0, 0)])
    visited = set()

    while queue:
        j1, j2 = queue.popleft()

        if j1 == target or j2 == target or j1 + j2 == target:
            return True

        next_states = [
            (jug1, j2),
            (j1, jug2),
            (0, j2),
            (j1, 0),
            (j1 - min(j1, jug2 - j2), j2 + min(j1, jug2 - j2)),
            (j1 + min(j2, jug1 - j1), j2 - min(j2, jug1 - j1))
        ]

        for state in next_states:
            if state not in visited:
                visited.add(state)
                queue.append(state)

    return False

print(canMeasureWater(7, 9, 4))
```

```python
from collections import deque

def canMeasureWater(jug1, jug2, target):
    # Check if the target is reachable by GCD conditions
    if target > jug1 + jug2 or target % math.gcd(jug1, jug2) != 0:
        return False

    # BFS to explore all possible states (amount of water in each jug)
    queue = deque([(0, 0)])  # Starting with both jugs empty
    visited = set([(0, 0)])

    while queue:
        j1, j2 = queue.popleft()

        # Check if we reached the target in either of the jugs
        if j1 == target or j2 == target or j1 + j2 == target:
            return True

        # List of possible next states:
        next_states = [
            (jug1, j2),  # Fill jug1
            (j1, jug2),  # Fill jug2
            (0, j2),      # Empty jug1
            (j1, 0),      # Empty jug2
            (j1 - min(j1, jug2 - j2), j2 + min(j1, jug2 - j2)),  # Pour water from jug1 to jug2
            (j1 + min(j2, jug1 - j1), j2 - min(j2, jug1 - j1))   # Pour water from jug2 to jug1
        ]

        # Add next states to queue if they haven't been visited yet
        for state in next_states:
            if state not in visited:
                visited.add(state)
                queue.append(state)

    return False  # If no solution is found

# Example Usage:
jug1Capacity = 7
jug2Capacity = 9
targetAmount = 4

# Check if we can measure the target amount of water
result = canMeasureWater(jug1Capacity, jug2Capacity, targetAmount)
print(result)  # Expected output: True
```
