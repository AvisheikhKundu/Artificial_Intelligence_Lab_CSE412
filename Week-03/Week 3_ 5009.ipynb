```python
graph = {
    '5' : ['3','7'],
    '3' : ['2', '4'],
    '7' : ['8'],
    '2' : [],
    '4' : ['8'],
    '8' : []
}

def dfs_iterative(graph, start_node):
    visited = set()
    stack = [start_node]
    while stack:
        node = stack.pop()
        if node not in visited:
            print(node)
            visited.add(node)
            stack.extend(reversed(graph[node]))
print("Following is the Depth-First Search (Iterative):")
dfs_iterative(graph, '5')
```

```python
from collections import deque
from collections import defaultdict

jug1, jug2, aim = 4, 3, 2

visited = defaultdict(lambda: False)


def waterJugSolver(amt1, amt2):
    if (amt1 == aim and amt2 == 0) or (amt2 == aim and amt1 == 0):
        print(amt1, amt2)
        return True

    if visited[(amt1, amt2)]:
        return False

    visited[(amt1, amt2)] = True
    print(amt1, amt2)

    return (
        waterJugSolver(0, amt2)
        or waterJugSolver(amt1, 0)
        or waterJugSolver(jug1, amt2)
        or waterJugSolver(amt1, jug2)
        or waterJugSolver(amt1 + min(amt2, (jug1 - amt1)), amt2 - min(amt2, (jug1 - amt1)))
        or waterJugSolver(amt1 - min(amt1, (jug2 - amt2)), amt2 + min(amt1, (jug2 - amt2)))
    )


print("Steps to reach the solution: ")
if not waterJugSolver(0, 0):
    print("No solution found.")
```

```python
graph = {
  '5': ['3', '7'],
  '3': ['2', '4'],
  '7': ['8'],
  '2': [],
  '4': ['8'],
  '8': []
}

visited = set()

def dfs(visited, graph, node):
    if node not in visited:
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)
        print(node)

print("Following is the Depth-First Search:")
dfs(visited, graph, '5')
```
