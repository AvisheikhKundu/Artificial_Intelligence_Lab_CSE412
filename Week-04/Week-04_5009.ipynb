```python
import csv
import heapq
import ipywidgets as widgets
from IPython.display import display, clear_output


# Class representing the graph
class Graph:
    def __init__(self, graph_dict=None, directed=True):
        self.graph_dict = graph_dict or {}
        self.directed = directed
        if not directed:
            self.make_undirected()

    # Method to make the graph undirected by adding reverse edges
    def make_undirected(self):
        for a in list(self.graph_dict.keys()):
            for (b, dist) in self.graph_dict[a].items():
                self.graph_dict.setdefault(b, {})[a] = dist

    # Method to connect two cities with a given distance
    def connect(self, A, B, distance=1):
        self.graph_dict.setdefault(A, {})[B] = distance
        if not self.directed:
            self.graph_dict.setdefault(B, {})[A] = distance

    # Method to get neighbors of a city
    def get(self, a, b=None):
        links = self.graph_dict.setdefault(a, {})
        if b is None:
            return links
        else:
            return links.get(b, None)


# Function to read data from a CSV file
def read_csv(filename):
    with open(filename, 'r') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip the header row
        data = [row for row in reader]
    return data


# Function to extract a graph from the read CSV data
def extract_graph(data):
    graph = Graph()
    for row in data:
        origin, destination, distance = row
        graph.connect(origin, destination, int(distance))
    return graph


# Heuristic function (optional)
def heuristic(city, end_city):
    return 0  # Modify this if using a heuristic


# A* search algorithm
def astar_search(graph, start, end):
    open_list = []  # Priority queue
    heapq.heappush(open_list, (0, start, []))  # Initialize the priority queue
    visited = set()  # Set to track visited nodes

    while open_list:
        cost, current, path = heapq.heappop(open_list)

        if current == end:  # If destination is reached
            path.append(current)
            return path, cost

        if current not in visited:
            visited.add(current)
            for neighbor, distance in graph.get(current).items():
                total_cost = cost + distance + heuristic(neighbor, end)
                heapq.heappush(open_list, (total_cost, neighbor, path + [current]))

    return None, float('inf')  # If no path is found


# Main function to create widgets and handle button clicks
def main():
    data = read_csv('indian-cities-dataset.csv')  # Use the correct file path
    graph = extract_graph(data)

    # Extract and sort unique cities from the dataset
    unique_cities = sorted(set(row[0] for row in data) | set(row[1] for row in data))
    cities_list = [None] + unique_cities  # Include None as the default option

    # Create widgets
    start_label = widgets.Label('Select Start City:')
    start_dropdown = widgets.Dropdown(options=cities_list)
    end_label = widgets.Label('Select Destination City:')
    end_dropdown = widgets.Dropdown(options=cities_list)
    calculate_button = widgets.Button(description='Calculate Route')
    output = widgets.Output()

    # Function to handle button click event
    def on_calculate_button_click(b):
        start_city = start_dropdown.value
        end_city = end_dropdown.value

        with output:
            clear_output(wait=True)
            if start_city is None or end_city is None:
                print("Error: Please select both start and destination cities.")
            elif start_city == end_city:
                print("Error: Start and destination cities cannot be the same.")
            else:
                # Call A* search algorithm and display the result
                path, total_distance = astar_search(graph, start_city, end_city)
                if path:
                    print("Shortest route from {} to {} is: {}".format(start_city, end_city, ' -> '.join(path)))
                    print("Total Path cost (in KM):", total_distance)
                else:
                    print("No path found.")

    # Attach the event handler function to the button click event
    calculate_button.on_click(on_calculate_button_click)

    # Display widgets
    display(start_label, start_dropdown, end_label, end_dropdown, calculate_button, output)


# Run the application
if __name__ == "__main__":
    main()
```
